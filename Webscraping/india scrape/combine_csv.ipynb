```python
# Rename columns in df2 and df3 to match df1's format
column_mapping = {
    'hsCode': 'HS CODE',
    'productDescription': 'PRODUCT DESCRIPTION',
    'consignee': 'CONSIGNEE',
    'consigneeAddress': 'CONSIGNEE ADDRESS',
    'countryOfOrigin': 'COUNTRY OF ORIGIN',
    'arrivalDate': 'ARRIVAL DATE',
    'quantity': 'QUANTITY',
    'unit': 'UNIT',
    'unitPriceUsd': 'UNIT PRICE IN USD'
}

# Apply the mapping to df2 and df3
df2_renamed = df2.rename(columns=column_mapping)
#df3_renamed = df3.rename(columns=column_mapping)

# Create missing columns with NaN values in df2 and df3
for col in df1.columns:
    if col not in df2_renamed.columns:
        df2_renamed[col] = pd.NA
    # if col not in df3_renamed.columns:
    #     df3_renamed[col] = pd.NA

# Keep only the columns that exist in df1
columns_to_keep = df1.columns

df2_filtered = df2_renamed[columns_to_keep]
# df3_filtered = df3_renamed[columns_to_keep]

# Combine all dataframes
combined_df = pd.concat([df1, df2_filtered], ignore_index=True)
```
